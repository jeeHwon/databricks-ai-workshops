```python
import mlflow

# Configuration — update these to match your environment
dbutils.widgets.text("experiment_name", "")
EXP_NAME = dbutils.widgets.get("experiment_name")  # User provides experiment name via widget

#Provide the catalog <schema> and <volume> details
CATALOG = "ai_workshop_series_catalog"
SCHEMA = "monash"
VOLUME = "mlflow_artifacts"

# Volume path must use the dbfs:/Volumes/ prefix
ARTIFACT_PATH = f"dbfs:/Volumes/{CATALOG}/{SCHEMA}/{VOLUME}"

mlflow.set_tracking_uri("databricks")
mlflow.set_registry_uri("databricks-uc")

# Create the experiment with the volume artifact location (only needs to run once)
if mlflow.get_experiment_by_name(EXP_NAME) is None:
    mlflow.create_experiment(name=EXP_NAME, artifact_location=ARTIFACT_PATH)

# Set the experiment as active — all traces will now go to the volume
mlflow.set_experiment(EXP_NAME)
```
